```python
'''↑のように行う前に基準となる列をデータフレームに追加しておくこと'''
'''最初のinput入力の際に検索したい列を入力すること'''


# データフレームから欲しい列を選択
# t : 欲しい列数
# find_column_df : 基準となる列を持っているデータフレーム
# df : 欲しい列があるデータフレーム
# new_df : 新しい空のデータフレーム（事前に作っておく）

def clean(t, find_column_df, df, new_df):

  # df と name の空リストを作成

  df_list = [ [] for i in range(0,t)]
  name_list = [ [] for i in range(0,t)]
  for i in range(0, t):

    # list に列名を入力し、追加

    n = input("抽出したい列名:")
    name_list[i].append(n)

    # df_list に列名と同じ列のdfを追加

    df_list[i].extend(df[n])

    # new_df に新しい df_list の列を追加

    new_df[i] = df_list[i]
  print(name_list)

  # 調べたいデータの列名を入力、 new_df に追加

  find_column = input("基準となる列名")
  new_df[find_column] = find_column_df[find_column]
  print(find_column)
  return new_df.head()

  # s : 基準データの行数分繰り返す 
  # t : 空リストの l の個数分繰り返す
  # T_df1_column : 基準となるデータフレームのカラム(列名)
  # T_df2_column : 検索したいデータフレームのカラム
  # new_df : 新しいデータフレーム（事前に準備）

def select_make(t , s, T_df1_column, T_df2_column, new_df):
  l = [ [] for i in range(t)]

  # T_df1_column(基準) の k番目が T_df2_column の中で一つでも等しかった場合、
  # df の m 列目で T_df1_column の k番目と T_df2_column が一致する要素を抽出
  # 等しいものが一つもなかったら、0を追加

  for k in range(0,s):
    for m in range(0, t):
      if(T_df1_column[k] == T_df2_column).any():
        l[m].extend(df[m][T_df1_column[k] == T_df2_column])
      else:
        l[m].append("0")

  # 空リストの個数分 new_df の i 番目を空リスト l の i 番目に入れる

  for i in range(0, t):
    new_df[i] = l[i]

  # new_df の0行目に0が入っている場合、除外する。 

  new_df_0 = new_df.index[new_df.iloc[:,0] == "0"]
  new_df = new_df.drop(new_df_0)
  return new_df.head()


def dup_reindex(df, new_df):

  # 0行目で重複したものを削除
  # indexの振り直し

  df_drop = df.drop_duplicates([0])
  new_df = df_drop.reset_index(drop = True)
  print(df.shape)
  print(new_df.shape)
  return new_df.head()
```

```python
japan1 = pd.DataFrame()

clean(4, df,japan, japan1)
```

```python
japan2 = pd.DataFrame()

select_make(4, df.shape[0], japan1["移動後の住所地（現住地）2020〜"], japan1[0], japan2 )
```

```python
japan3 = pd.DataFrame()

dup_reindex(japan2, japan3)
```
